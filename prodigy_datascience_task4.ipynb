```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'C:/Users/Udhaya/Documents/prodigy/twitter_training.csv'
data = pd.read_csv(file_path, encoding='ISO-8859-1')
```

```python
data.columns = ['tweet_id', 'entity', 'sentiment', 'text']
```

```python
data = data.dropna(subset=['text'])
```

```python
print(data.head())
```

```python
plt.figure(figsize=(8,6))
sns.countplot(data['sentiment'], palette='coolwarm')
plt.title('Sentiment Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='entity', hue='sentiment', palette='coolwarm')
plt.title('Sentiment per Entity')
plt.xticks(rotation=45)
plt.show()
```
